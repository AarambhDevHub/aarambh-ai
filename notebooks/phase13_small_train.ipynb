```python
!cargo build --release -p aarambh-ai --features cuda
```

```python
!mkdir -p checkpoints/wikitext103_small && cargo run --release -p aarambh-ai --features cuda -- train --config configs/wikitext103_small.toml 2>&1 | tee checkpoints/wikitext103_small/train.log
```

```python
%%bash
MODEL=$(python3 -c 'import json, pathlib; p=json.load(open("checkpoints/wikitext103_small/latest.json"))["path"]; print(pathlib.Path(p) / "model.safetensors")')
cargo run --release -p aarambh-ai --features cuda -- infer \
  --config configs/wikitext103_small.toml \
  --model "$MODEL" \
  --tokenizer checkpoints/wikitext103_small/tokenizer.json \
  --prompt "To be, or not to be" \
  --max-tokens 64 \
  --greedy \
  --stream
```
